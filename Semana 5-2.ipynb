```python
#if
#Problema: dado el precio de 3 productos ingresados por el teclaso,aplicar un descuento del 10% si la
#venta es mayor o igual a 1000
```

```python
#solucion
P1=float(input("Producto 1: "))
P2=float(input("Producto 2: "))
P3=float(input("Producto 3: "))
total=P1+P2+P3
if total >=1000:
  total *=0.9
print(f"El total a pagar es. {total}")
```

```python
ventas = float(input("importe de ventas realizadas: "))
if ventas <0:
  print("Datos erroneos")
elif ventas <=1000:
  print("categoria A")
elif ventas<=2000:
  print("Categoria B")
elif ventas <=5000:
  print("Categoria C")
else:
  print("Categoria D")
```

```python
#Problema 01:
#Mostrar los elementoa de una lista utilizando for
```

```python
#Solucion
aula=['Barrios','Sevillano','Zuñiga','Paye']
#por cada alumno que esta en la lista, imprimir alumno
for alumno in aula:
  print(alumno)
```

```python
#Problema 02:
#dado una cadena ingresado por teclado,
#mostrar cada caracter de la cadena utilizando la estructura for
```

```python
#solucion
cadena=input("Ingrese cadena:")
for caracter in cadena:
  print(caracter)
```
