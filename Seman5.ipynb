1.COLECIONES DE DATOS

```python
#1.1 Listas
#Es una coleccion de elementos que puede ser ordenanda, modificada, etc...
#Se identifica por los corchetes
#ejemplo:
aula = ['Huertas','Barvo','Bonifacio','Caballero']
print(aula)
```

```python
#Tambien se puede imprimir con solo menionar la Lista
aula
```

```python
#Podemos extraer cada elemento por su posicion
print(aula[0])
print(aula[1])
```

```python
#Puedo asignar un valor a una posicion de la lista
aula[1] ='Fernandez'
aula[1]
```

```python
#Mostramos como quedo la lista
aula
```

```python
#para agregar un elemento utilizamos append
aula.append("Mamani")
aula.append("Bonifacio")
aula
```

```python
#Para eliminar elemento utilizamos remove[]
aula.remove("Huertas")
aula
```

```python
#Si hubiera varios elementos de valore siguales se elimina la primera que coinciden
aula.remove("Bonifacio")
aula
```

```python
#para obtener el tama√±o de la lista utilizamos len()
len(aula)
```

```python
#para recorrer podemos utilizar una estructura repetitiva
for alumno in aula:
  print(alumno)
```

```python
#Para invertir la lista utilizamos reverse()
#Invertir no siginifica ordenar en forma descendente sin invertir la lista
aula.reverse()
aula
```

```python
#Para ordenar una lista utilizamos sort()
aula.sort()
aula
```

```python
#1.3 Conjuntos
#Es una coleccion de elementos que no esta ordenado
#significa que no tiene un indice
conjunto_aula={'Huertas','Bravo','Bonifacion','Caballero'}
conjunto_aula
```

```python
#Se puede recorrer utilizando una estructura repetitiva
#ejemplo: for
for alumno in conjunto_aula:
  print(alumno)
```

```python
#Observacion: No se puede acceder por posicion pues no considera indexacion
#por ejemplo:
conjunto_aula[2] devolveria un mensaje de error#examen
```

```python
#Para saber la cantidad de elementos utilizamos len()
len(conjunto_aula)
```

```python
#Para agregar utilizamos la funcion add()
conjunto_aula.add("Paye")
conjunto_aula
```

```python
#para eliminar elementos utilizaos la funcion remove()
conjunto_aula.remove("Bravo")
conjunto_aula
```

```python
#1.4 Diccionarios
#Es una coleeccion de elementos. que estan indexados,
#no estan ordenados y se pueden modificar
#son escritos entre llavez y estan formados por  pares de elementos
#INDICE: VALOR
diccionario_aula={2:'Huertas','1':'Bravo','dos':'Bonifacion','3':'Caballero','0':'Paye'}
diccionario_aula
```

```python
#Del diccionario: diccionario_aula extraer el valor cuyo indice sea cero en string
diccionario_aula['0']
```

```python
diccionario_aula[2]
```

```python
diccionario_aula["1"]
```

```python
diccionario_aula["dos"]
```

```python
#Para a√±adir un par de elementos al diccionario hacemos:
diccionario_aula['20']="Velez"
diccionario_aula
```

```python
#Para eliminar un valor se utiliza pop()
#Por ejemplo, eliminar el elemento de indice dos. ejemplo: Para eliminar a Huertas
diccionario_aula.pop(2)
diccionario_aula
```

```python
#Tambienb se puede eliminar utilizando del()
del(diccionario_aula['dos'])
diccionario_aula
```

```python
#Para obtener el indice y el valor de cada elemento utilizamos items()
for indice,valor in diccionario_aula.items():
  print(indice,valor)
```
